```python
import numpy as np

def determinant(matrix):
    try:
        det = np.linalg.det(matrix)
        return det
    except np.linalg.LinAlgError:
        print("The matrix is singular, and the determinant cannot be calculated.")

#I will input my matrix here
matrix = np.array([[1, 0, 1],
                   [0, 0, 1],
                   [1, 0, 1]])

result = determinant(matrix)

if result is not None:
    print(f"Determinant of the matrix:\n{matrix}\nis: {result}")
```
